```python
import pandas as pd

metadata = pd.read_csv('metadata.csv')

metadata.head()
```

```python
# Add "content/" prefix to every filepath in the metadata DataFrame
metadata['filepath'] = 'content/' + metadata['filepath']

# Check if 'content_type' column exists before dropping it
metadata = metadata.drop(columns=['content_type'])

# Display the first few rows to verify the change
metadata.head()
```

```python
import os

# Create a new column to track if the file exists
metadata['file_exists'] = metadata['filepath'].apply(lambda x: os.path.exists(x))

# Filter the DataFrame to keep only rows where the file exists
metadata = metadata[metadata['file_exists']]

# Drop the temporary column
metadata = metadata.drop(columns=['file_exists'])
metadata.reset_index(drop=True, inplace=True)

# Display the number of rows after filtering
print(f"Number of files found: {len(metadata)}")
metadata.head()

```

```python
import re

def remove_markdown_images(content):
    """
    Remove markdown image syntax from content.
    This handles both inline images ![alt text](url) and reference style images.
    """
    # Remove inline images ![alt text](url)
    content = re.sub(r'!\[.*?\]\(.*?\)', '', content)
    
    # Remove reference style images ![alt text][id]
    content = re.sub(r'!\[.*?\]\[.*?\]', '', content)
    
    return content

def process_file(filepath):
    """Process a single file to remove markdown images"""
    try:
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read()
            print(content)
        
        # Remove markdown images
        new_content = remove_markdown_images(content)
        
        # Write the modified content back to the file
        with open(filepath, 'w', encoding='utf-8') as file:
            file.write(new_content)
            
        return True
    except Exception as e:
        print(f"Error processing {filepath}: {str(e)}")
        return False

# Process all files in the metadata DataFrame
processed_count = 0
for filepath in metadata['filepath']:
    if process_file(filepath):
        processed_count += 1

print(f"Processed {processed_count} files out of {len(metadata)}")
print(f"Removed markdown images from all files")
```

```python
# Save the processed metadata back to a CSV file
metadata.to_csv('metadata.csv', index=False)
print(f"Metadata saved to metadata.csv with {len(metadata)} records")
```
